## Temporal features

```python
datetimes = pd.to_datetime(df['created'], format='%d%b%Y %H:%M:%S')

temporal_data = pd.DataFrame()
```

#### Add $\Big(\sin\left(\frac{2\pi}{52}\textrm{week_number}\right),  \cos\left(\frac{2\pi}{52}\textrm{week_number}\right)\Big)$

```python
temporal_data['week_number_sine'] = [np.sin((2*np.pi/52)*date.week) for date in datetimes]
temporal_data['week_number_cosine'] = [np.cos((2*np.pi/52)*date.week) for date in datetimes]
```

#### Add $\Big(\sin\left(\frac{2\pi}{12}\textrm{month}\right),  \cos\left(\frac{2\pi}{12}\textrm{month}\right)\Big)$

```python
temporal_data['month_sine'] = [np.sin((2*np.pi/12)*date.month) for date in datetimes]
temporal_data['month_cosine'] = [np.cos((2*np.pi/12)*date.month) for date in datetimes]
```

#### Add year

```python
temporal_data['year'] = [date.year for date in datetimes]
```

## Numerical features

#### Median Imputation
Instead of mean use median imputation to fill missing data since median is robust to outliers

```python
numerical_data = df[numerical_features].fillna(df[numerical_features].median())
```

## Categorical features

There are two ways to manipulate these features:

#### Keep nan as a separate class and use 3-dimensional one-hot encoding vector

```python
categorical_data = pd.get_dummies(
    df[categorical_features], columns=categorical_features, dummy_na=True
).rename(
    lambda col: col.replace('.0', ''), axis=1
)
```

#### Assign 0 to nan and do not use one-hot encoding for categories with binary values

```python
categorical_data = df[categorical_features].fillna(value=0)
```

```python
categorical_data.loc[categorical_data['a_gender2']=='male', 'a_gender2'] = 0
categorical_data.loc[categorical_data['a_gender2']=='female', 'a_gender2'] = 1
categorical_data['a_gender2'] = categorical_data['a_gender2'].astype(np.float64)
```

## Remaining features

```python
remaining_data = []
```

```python
# labs_mrdt_before
temp_df = df['labs_mrdt_before'].copy()
temp_df.loc[temp_df.isna()] = 'none'
temp_df = pd.get_dummies(temp_df, prefix='labs_mrdt_before')
remaining_data.append(temp_df)
```

```python
# s_muac
muac_mapping = {
    np.nan: 0.0,
    'green': 1.0,
    'yellow': 2.0,
    'orange': 3.0,
    'red' : 4.0
}

temp_df = df['s_muac'].copy()
temp_df[temp_df.isna()] = np.nan
temp_df = temp_df.apply(lambda x: muac_mapping[x])
remaining_data.append(temp_df.to_frame())
```

```python
# vaccines
temp_df = df['vaccines'].copy()
temp_df[temp_df.isna()] = 'notime'
temp_df = pd.get_dummies(temp_df, prefix='vaccines')
remaining_data.append(temp_df)
```

```python
remaining_data = pd.concat(remaining_data, axis=1)
```

## Normalization

```python
spatial_data = (spatial_data-spatial_data.mean())/spatial_data.std()
temporal_data = (temporal_data-temporal_data.mean())/temporal_data.std()
numerical_data = (numerical_data-numerical_data.mean())/numerical_data.std()
```
